```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/kmeans/student_clustering.csv')
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,12):
   km = KMeans(n_clusters = i)
   km.fit_predict(df)
   wcss.append(km.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,12),wcss)
```

```python
km = KMeans(n_clusters=4)

y = km.fit_predict(df)
```

```python
y
```

```python
df[y == 0]
```

```python
df[(y == 1) | (y == 2)]
```

```python
# Scatter plot the clusters
# Separate row selection and column selection using .iloc
plt.scatter(df.iloc[y == 0, 0], df.iloc[y == 0, 1], color = 'blue', label = 'Cluster 1')
plt.scatter(df.iloc[y == 1, 0], df.iloc[y == 1, 1], color = 'red', label = 'Cluster 2')
plt.scatter(df.iloc[y == 2, 0], df.iloc[y == 2, 1], color = 'green', label = 'Cluster 3')
plt.scatter(df.iloc[y == 3, 0], df.iloc[y == 3, 1], color = 'purple', label = 'Cluster 4')
```
